```python
import sarpy
import sarpy.analysis.special_finder as sp
reload(sarpy.analysis.special_finder)
import sarpy.analysis.gather_data as gd
import pandas
import collections
import os

def getT1data(day0scans):
    
    datat1 = []
    
    if day0scans:

        for t in day0scans:
        
            t1 = sarpy.Scan(t).adata['T1_LL'].data
            roi = sarpy.Scan(t).adata['T1_LL_roi'].data
        
            t1roi = t1*roi
            t1roi = t1roi[numpy.isfinite(t1roi)]
                      
        datat1.append(list(t1roi.flatten()))
    
    return [x for x in flatten(datat1)]
                  
```

```python
## First make a df from the available masterlists

masterlists = ['HerP2','HerS10','HerS11','NecS1','NecS3']

dfs = pandas.DataFrame()

for m in masterlists:
    dfs = dfs.append(gd.df_from_masterlist(m))
    
#dfs

exclude_patients = ['HerS10Do01','HerS10Do02','HerS10Do03','HerS10Do04','HerS10Do05',
                                         'HerS11Bs08','HerS11Bs09','HerS11Bs10','HerS11Bs11','HerS11Bs08'
                                         'HerP2Bs06','HerP2Bs07','HerP2Bs08']

for i in exclude_patients:
    
    try:
        dfs.ix[i] = numpy.nan
    except:
        continue

#HerP2 and HerS11 can be combined B
#HerS10 is orthotopic D
# NecS1 and NecS3 is Hs
```
