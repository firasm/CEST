```python
import pandas
import sys
import os
import collections
import sarpy.analysis.gather_data as gd

def dfarray(df,keyList):
    
    newarr = []
    for key in keyList:
        newarr.append(df[key].values)        
    return newarr
```

### T1 maps at 0h

```python
## First make a df from the available masterlists

masterlists = ['HerP2','HerS10','HerS11','NecS1','NecS3']

dfs = pandas.DataFrame()

for m in masterlists:
    dfs = dfs.append(gd.df_from_masterlist(m))
```

```python
csvlist = collections.OrderedDict()

csvlist={'HerP2' : os.path.join(os.path.expanduser('~'),'sdata','HerP2','export','LL-0h_T1_LL.csv'),
         'HerS10' : os.path.join(os.path.expanduser('~'),'sdata','HerS10','export','LL-0h_T1_LL.csv'),
         'HerS11' : os.path.join(os.path.expanduser('~'),'sdata','HerS11','export','LL-0h-_T1_LL.csv'),
         'NecS1' : os.path.join(os.path.expanduser('~'),'sdata','NecS1','export','LL-0h_T1_LL.csv'),
         'NecS3' : os.path.join(os.path.expanduser('~'),'sdata','NecS3','export','LL-0h-_T1_LL.csv')
        }
T1 = pandas.DataFrame()
    
for k,v in csvlist.iteritems():

    T1 = T1.append(pandas.read_csv(v,skipfooter=5,index_col = 'Animal ID'))
T1 = T1.merge(dfs, on=T1.index,how='left')
```

```python
# Average of tumour types
tumourtypes = ['B','D','H']

for t in tumourtypes:

    print('Tumour Type {0}: \n \t animals: {1}, \n \t T1 mean: {2} ms, \n \t std: {3} ms'.format(t,
                                                    T1[T1['tumourtype']==t]['Average'].count(), 
                                                    T1[T1['tumourtype']==t]['Average'].mean(),
                                                    T1[T1['tumourtype']==t]['Average'].std()))
```

```python
#Average of tumour types
pos = ['s','o']

for p in pos:

    print('Tumour Location {0}: \n \t animals: {1}, \n \t T1 mean: {2} ms, \n \t std: {3} ms'.format(p,
                                                    T1[T1['location']==p]['Average'].count(), 
                                                    T1[T1['location']==p]['Average'].mean(),
                                                    T1[T1['location']==p]['Average'].std()))
```
