```python
import numpy
import pylab
import sarpy
import sarpy.fmoosvi.analysis
import sarpy.fmoosvi.getters as getters
import scipy.optimize
```

```python
testScan = sarpy.Scan('HerP2Bs07.jU1/5')
```

```python
oldT1map = testScan.adata['T1map_LL'].data
pylab.imshow(oldT1map[7:35,48:95,3],vmin = 500, vmax = 3000, cmap = 'hot')
pylab.colorbar()
```

```python
newT1map,fr = sarpy.fmoosvi.analysis.h_fit_T1_LL_FAind(testScan, bbox = numpy.array([7,35,48,95]))
```

```python
T1eff = empty_like(newT1map)

for slice in xrange(T1eff.shape[-1]):
    for y in xrange(T1eff.shape[-2]):
        for x in xrange(T1eff.shape[-3]):
            try:
                T1eff[x,y,slice] = fr[x,y,slice]['fit_params'][2]
            except:
                T1eff[x,y,slice] = numpy.nan

fig = pylab.figure(figsize=(10,5))
pylab.subplot(1,2,1)
im=pylab.imshow(T1eff[7:35,48:95,3],vmin = 0, vmax = 2000, cmap = 'hot', interpolation='none')
pylab.title('T1_effective')
#colorbar_ax = fig.add_axes([0.95, 0.1, 0.05, 0.8])
#fig.colorbar(im, cax=colorbar_ax)

pylab.subplot(1,2,2)
im2=pylab.imshow(newT1map[7:35,48:95,3],vmin = 100, vmax = 2000, cmap = 'hot', interpolation='none')
pylab.title('T1-proper')
colorbar_ax2 = fig.add_axes([0.95, 0.25, 0.03, 0.5])
fig.colorbar(im2, cax=colorbar_ax2)
```

### Sample sucessful fit

```python
raw = numpy.squeeze(sarpy.io.BRUKERIO.fftbruker(testScan.fid))
dat=numpy.fliplr(numpy.flipud(numpy.transpose(raw.reshape(128,64,30,8),[1,0,3,2])))
test = dat[20,58,3]
inversion_time = testScan.method.PVM_InversionTime
tao = testScan.method.PVM_RepetitionTime

total_TR=testScan.method.Inv_Rep_time

n_data = numpy.linspace(0,29,30).astype(int)

params = [1e7, -100, 1000, 0]

params[0] = numpy.real(numpy.mean(test[-5:]))
params[1] = numpy.real(numpy.divide(-test[0],params[0]))
params[2] = 1200
params[3] = numpy.angle(numpy.mean(test[-5:]))

fit_params,cov,infodict,mesg,ier = scipy.optimize.leastsq(sarpy.fmoosvi.analysis.h_residual_T1_FAind,params,args=(test,tao,n_data), full_output=True, maxfev = 200)
fit = sarpy.fmoosvi.analysis.h_func_T1_FAind(fit_params,tao, n_data)
figsize(10,5)
pylab.subplot(1,2,1)
pylab.plot(n_data*tao,numpy.real(test),'ro')
pylab.plot(n_data*tao,numpy.real(fit),'r')
pylab.subplot(1,2,2)
pylab.plot(n_data*tao,numpy.imag(test),'g+')
pylab.plot(n_data*tao,numpy.imag(fit), 'g')
print 'initial guess: ',params[2],'fit result: ', fit_params[2]
print 
print mesg
```
