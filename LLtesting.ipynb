### Manual Fitting

```python
import sarpy
import sarpy.fmoosvi.getters as getters
import scipy.stats
import sarpy.fmoosvi.analysis as an

LL=sarpy.Experiment('HerS11Bs09').find_scan_by_protocol('.*LL.*')[0]
scn = LL

x = sarpy.io.BRUKERIO.fftbruker(scn.fid)
tau = scn.method.PVM_RepetitionTime
num_slices = getters.get_num_slices(scn.dirname,0)                                        
t1points = numpy.divide(x.shape[-2],num_slices)     
data=numpy.fliplr(
     numpy.flipud(numpy.transpose(x.reshape(x.shape[0],
                                            x.shape[1],
                                            t1points,
                                            num_slices),
                                            [1,0,3,2])))
Nframes = scn.method.Nframes
t_data = numpy.arange(0,Nframes) * tau + scn.method.PVM_InversionTime
```

```python
idx = (48,75,5)

y_data = data[idx]
x_data = t_data

tau = scn.method.PVM_RepetitionTime

params = numpy.ones(4)                   
y_data = data[idx]

fit_dict = {}
fit_algorithm = 'leastsq'

## Guesses at parameters 
params[3] = numpy.angle(numpy.mean(y_data[-5:]))  #phi
params[0] = numpy.abs(numpy.mean(y_data[-5:])) #a
params[1] = numpy.real(numpy.divide(y_data[0]*numpy.exp(-1j*params[3]),params[0])) #b

# this is to catch divide by 0 errors
if params[3] == 0: 
    params[3] = numpy.angle(numpy.mean(y_data[-10:]))

    if params[3] == 0:   
        params[3] = 0.5 #random starting value
        

# Getting a good guess for T1eff by finding the zero crossing
yp = y_data[0:5]
xp = t_data[0:5]

A = numpy.array([xp, numpy.ones(xp.shape[0])])
w=numpy.linalg.lstsq(A.T,yp)[0]
params[2] = numpy.real(numpy.divide(-w[1],w[0])) 

if fit_algorithm == 'leastsq':

    fit_params,cov,infodict,mesg,ier = scipy.optimize.leastsq(
                                            an.h_residual_T1_FAind,
                                            params,
                                            args=(y_data,t_data), 
                                            full_output = True,
                                            maxfev = 200)

    # Flip the sign to try and make it fit again
    if ier > 4: #This is to try and refit 
        
        print ('retry')
        params[0] = numpy.abs(numpy.mean(y_data[-5:])) +numpy.pi/4
        #params[3] = -numpy.angle(numpy.mean(y_data[-5:]))+numpy.pi/4 #phi

        fit_params,cov,infodict,mesg,ier = scipy.optimize.leastsq(
                                                an.h_residual_T1_FAind,
                                                params,
                                                args=(y_data,t_data), 
                                                full_output = True,
                                                maxfev = 200)
print params
print fit_params
print ier,mesg
```
