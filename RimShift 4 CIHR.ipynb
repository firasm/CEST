Rim Shift for just one profile
==============================

```python
import sarpy
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import PolyCollection
from matplotlib.colors import colorConverter
import matplotlib.pyplot as plt
import numpy as np
def rim_view(patname = 'herS3M06', slice=3, row=42, rangestart=70, rangeend=80, dceprotocol='.*DCE',
             startframe=0, endframe=-1, stepframe=2):
    expt = sarpy.Patient(patname).studies[0]
    dcescan = expt.find_scan_by_protocol(dceprotocol)
    dcedata = dcescan[0].pdata[0].data
    pylab.figure(figsize=(20,10))
    pylab.subplot(1,4,1)
    pylab.imshow(dcedata[:,:,slice,20])
    pylab.plot([rangestart,rangeend],[row,row],'-')
    pylab.title(patname)
    # plot profiles through one position in the image
    pylab.subplot(1,4,2)
    if endframe== -1:
        endframe=dcedata.shape[-1]
    for i in arange(startframe, endframe, stepframe):
        plot(dcedata[row,rangestart:rangeend,slice,i])
    pylab.xlabel('position [pixels]')
    pylab.ylabel('intensity [a.u.]')
    
    # determine the shift in peak position through centroiding of the peak
    peak_pos = []
    pylab.subplot(1,4,3)
    for i in arange(startframe, endframe, 1):
        peak_pos.append(sum(dcedata[row,rangestart:rangeend,slice,i]*arange(rangeend-rangestart))/sum(dcedata[row,rangestart:rangeend,slice,i]))
    pylab.plot(peak_pos)
    pylab.xlabel('time [frame]')
    pylab.ylabel('peak shift [pixels]')
    ax = pylab.subplot(1,4,4, projection='3d')
    ax.view_init(50, 270)

    xs = arange(startframe, endframe)
    ys = arange(rangestart, rangeend)
    zs = dcedata[row,rangestart:rangeend,slice,startframe:endframe]
    
    cc = lambda arg: colorConverter.to_rgba(arg, alpha=0.6)
    verts = []
    for frame in xrange(startframe, endframe):
        verts.append(list(zip(ys, dcedata[row, rangestart:rangeend, slice, frame])))
    poly = PolyCollection(verts, facecolors = [cc('r'), cc('g'), cc('b'),
                                               cc('y')])
    poly.set_alpha(0.7)
    ax.add_collection3d(poly, zs=xs, zdir='y')

    ax.set_xlabel('pixelpos')
    ax.set_xlim3d(rangestart, rangeend)
    ax.set_ylabel('frame')
    ax.set_ylim3d(startframe, endframe)
    ax.set_zlabel('intensity')
    ax.set_zlim3d(numpy.min(zs), numpy.max(zs))

    pylab.show()

    
    return(verts, xs)
```

```python
(v,xs) = rim_view(patname='herS3M01', slice=3, row=28, rangestart=60, rangeend=80)
```

now that we have all these profiles we need to find how much they 'move'
-------------------------------------------------------------------------

```python
startframe = 0
endframe = n_frames
peak_pos = empty([n_frames, len(rim_x)], dtype='float')

# profiles = empty([n_frames, len(rim_x),len(t)], dtype='float')

ctr=0
for prof in xrange(len(rim_x)):
    for i in arange(startframe, endframe, 1):
        # the centroiding method to find whether a profile shifts
        peak_pos[i,ctr]=sum(profiles[i,prof,:]*arange(len(t))/sum(profiles[i,prof,:]))
    ctr+=1

baseline_pos = numpy.mean(peak_pos[0:10,:],axis=0)
postline_pos = numpy.mean(peak_pos[20:30,:],axis=0)
rimshift = postline_pos - baseline_pos


# the steepest gradient finder of the profile edge
#rimshift = numpy.empty([len(rim_x)])
#ctr=0
#for prof in xrange(len(rim_x)):
#    gr1 = abs(profiles[2,prof,:-1]-profiles[2,prof,1:])
#    gr2 = abs(profiles[20,prof,:-1]-profiles[20,prof,1:])
#    rimshift[ctr] = sum(gr1*arange(len(gr1))) / sum(gr1) - sum(gr2*arange(len(gr2))) / sum(gr2)
#    ctr+=1

        
```
