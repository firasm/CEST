```python
# Get unique patients

pats= set()
for studies in hpg.studies:
    pats.add(studies.subject.SUBJECT_id)
```

```python
dates = {}

for p in pats:
    stds = sarpy.Patient(p).studies
    curr = []
    
    for s in stds:
        t = string.join(s.subject.DATE.split(' '))
        curr.append(re.match('(.{3})\s(.{3})\s+([0-9]+)',t).group())
    dates[p]=curr
```
