```python
import sarpy
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
def rim_view(patname = 'herS3M06', slice=3, row=42, rangestart=70, rangeend=80, dceprotocol='.*DCE',
             startframe=0, endframe=-1, stepframe=2):
    expt = sarpy.Patient(patname).studies[0]
    dcescan = expt.find_scan_by_protocol(dceprotocol)
    dcedata = dcescan[0].pdata[0].data
    figure(figsize(20,10))
    subplot(1,4,1)
    imshow(dcedata[:,:,slice,20])
    plot([rangestart,rangeend],[row,row],'-')
    pylab.title(patname)
    # plot profiles through one position in the image
    subplot(1,4,2)
    if endframe== -1:
        endframe=dcedata.shape[-1]
    for i in arange(startframe, endframe, stepframe):
        plot(dcedata[row,rangestart:rangeend,slice,i])
    pylab.xlabel('position [pixels]')
    pylab.ylabel('intensity [a.u.]')
    
    # determine the shift in peak position through centroiding of the peak
    peak_pos = []
    subplot(1,4,3)
    for i in arange(startframe, endframe, 1):
        peak_pos.append(sum(dcedata[row,rangestart:rangeend,slice,i]*arange(rangeend-rangestart))/sum(dcedata[row,rangestart:rangeend,slice,i]))
    plot(peak_pos)
    pylab.xlabel('time [frame]')
    pylab.ylabel('peak shift [pixels]')
    subplot(1,4,4, projection='3d')
    xs = tile(arange(startframe, endframe),rangeend-rangestart)
    ys = repeat(arange(rangestart, rangeend),endframe-startframe)
    zs = dcedata[row,rangestart:rangeend,slice,startframe:endframe]
    scatter(xs,ys,zs)
```

```python
rim_view(patname='herS3M01', slice=3, row=28, rangestart=60, rangeend=80)
```
