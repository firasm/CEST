# Control Animals (NecS1)

```python
avg = numpy.array(dfdataC['0h-T1map_LL_avg'].tolist())
weights= numpy.array(dfdataC['0h-T1map_LL_px'].tolist())
avg0c = numpy.average( avg[numpy.isfinite(avg)], weights = weights[numpy.isfinite(weights)])
std0c = numpy.std(avg[numpy.isfinite(avg)])
```

```python
avg = numpy.array(dfdataC['24h-T1map_LL_avg'].tolist())
weights= numpy.array(dfdataC['24h-T1map_LL_px'].tolist())
avg1c = numpy.average( avg[numpy.isfinite(avg)], weights = weights[numpy.isfinite(weights)])
std1c = numpy.std(avg[numpy.isfinite(avg)])
```

```python
# Subtract the means and use the errors added in quadrature

print(numpy.str(avg1c - avg0c) + ' +/- ' + numpy.str(numpy.sqrt( std0c**2 + std1c**2)))
```
