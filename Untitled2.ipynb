```python
import pylab
```

```python
times = [-0.5, -3,-8, -18, -10]
         
abst = [1024, 1935, 5795, 2472, 6658]

year = [2009, 2010, 2011, 2012, 2013]
```

```python
figsize(14,7)
ax=pylab.subplot(121)
plot(year, abst,'*', markersize=30)
axhline(linewidth=4, color='r',y=8000)
ax.tick_params(labelsize=12,direction='out', length=6,pad=12)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_xticklabels([2008, 2009, 2010, 2011, 2012, 2013,2014, 2015])
xlim(2008,2015)
xlabel('\n ISMRM Year',size=14)
ylim(500,8500)
ylabel('Abstract # ',size=14)
title('Abstract Start History\n',size=18)


ax=pylab.subplot(122)
plot(year, times,'*', markersize=30)
axhline(linewidth=4, color='r')
ax.tick_params(labelsize=12,direction='out', length=6,pad=12)

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_xticklabels([2008, 2009, 2010, 2011, 2012, 2013,2014, 2015])
xlim(2008,2015)
xlabel('\n ISMRM Year',size=14)
ylim(-20,5)
ylabel('Minutes from deadline',size=14)
title('Abstract Finish History\n',size=18)
```
