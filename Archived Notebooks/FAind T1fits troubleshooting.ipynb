```python
import sarpy
import numpy
import scipy
import pylab
import pprint

import sarpy.fmoosvi.analysis

pp = pprint.PrettyPrinter(indent=4)
```

# Testing new FA independent fits

```python
testScan = sarpy.Scan('HerP2Bs07.jU1/5')
raw = numpy.squeeze(sarpy.io.BRUKERIO.fftbruker(testScan.fid))
dat=numpy.fliplr(numpy.flipud(numpy.transpose(raw.reshape(128,64,30,8),[1,0,3,2])))
test = dat[20,71,6]
inversion_time = testScan.method.PVM_InversionTime
tao = testScan.method.PVM_RepetitionTime

total_TR=testScan.method.Inv_Rep_time

n_data = numpy.linspace(0,29,30).astype(int)
params = [1e7, -100, 1000, 5.1]

fit_params,cov,infodict,mesg,ier = scipy.optimize.leastsq(sarpy.fmoosvi.analysis.h_residual_T1_FAind,params,args=(test,tao,n_data), full_output=True, maxfev = 500)
fit = sarpy.fmoosvi.analysis.h_func_T1_FAind(fit_params,tao, n_data)
figsize(10,5)
pylab.subplot(1,2,1)
pylab.plot(n_data*tao,numpy.real(test),'ro')
pylab.plot(n_data*tao,numpy.real(fit),'r')
pylab.subplot(1,2,2)
pylab.plot(n_data*tao,numpy.imag(test),'g+')
pylab.plot(n_data*tao,numpy.imag(fit), 'g')
print fit_params
print params
print mesg
```

# Sample Fit ( FA-dependent)

```python
testScan = sarpy.Scan('HerP2Bs07.jU1/5')
fid = numpy.squeeze(sarpy.io.BRUKERIO.fftbruker(testScan.fid))
dat=numpy.fliplr(numpy.flipud(numpy.transpose(fid.reshape(128,64,30,8),[1,0,3,2])))
test = numpy.real(dat)[25,35,6]
inversion_time = testScan.method.PVM_InversionTime
repetition_time = testScan.method.PVM_RepetitionTime

t_data = numpy.linspace(inversion_time,testScan.pdata[0].data.shape[3]*repetition_time,testScan.pdata[0].data.shape[3])

params = [3E5, 2, 350, 5.0]

fit_params,cov,infodict,mesg,ier = scipy.optimize.leastsq(sarpy.fmoosvi.analysis.h_residual_T1,params,args=(test,t_data), full_output=True, maxfev = 200)
fit = sarpy.fmoosvi.analysis.h_func_T1(fit_params,t_data)
pylab.plot(t_data,numpy.real(test),'o')
pylab.plot(t_data,fit)
```
