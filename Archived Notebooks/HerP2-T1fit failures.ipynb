# Sample Fits

```python
testScan = sarpy.Scan('HerP2Bs07.jU1/5')
fid = numpy.squeeze(sarpy.io.BRUKERIO.fftbruker(testScan.fid))
dat=numpy.fliplr(numpy.flipud(numpy.transpose(fid.reshape(128,64,30,8),[1,0,3,2])))
test = numpy.real(dat)[25,35,6]
inversion_time = testScan.method.PVM_InversionTime
repetition_time = testScan.method.PVM_RepetitionTime

t_data = numpy.linspace(inversion_time,testScan.pdata[0].data.shape[3]*repetition_time,testScan.pdata[0].data.shape[3])

params = [3E5, 2, 350, 5.0]

fit_params,cov,infodict,mesg,ier = scipy.optimize.leastsq(sarpy.fmoosvi.analysis.h_residual_T1,params,args=(test,t_data), full_output=True, maxfev = 200)
fit = sarpy.fmoosvi.analysis.h_func_T1(fit_params,t_data)
pylab.plot(t_data,numpy.real(test),'o')
pylab.plot(t_data,fit)
```
